## Lists

Lists are a type of object in Python that are basically simple containers for other objects. They are one of the most fundamental data structures in programming. Used properly, they can be very powerful. 

```python
my_list = ['apples','oranges','grapes']
```

Lists keep items in their original order. You can get an item in a list by "slicing" the list using the position number of that item in that order. Position numbers start at 0. 

```python
second_list_item = my_list[1]

print(second_list_item)
```

You can replace an item in a list by reassigining that slice of the dictionary to a new object

```python
my_list[1] = 'pineapples'

my_list
```

You can add items to a list by using the "append" method. 

```python
new_fruit = 'mango'

my_list.append(new_fruit)

my_list
```

The key-value pair system allows us to retrieve the values using keys, without having to know the location of the item in the dictionary. (Dicionaries have no sense of order)

```python
color_dict['apple']
```

# Loops

In python, you have the ability to cycle through a set of objects if they are part of a valid "iterable"

An iterably is basically anything that can be iterated over, i.e it is made up of smaller components that can be read one by one. 

Lists are an iterable, and they are very frequently iterated over. 



```python
my_list = ['apples','oranges','grapes']

for fruit in my_list:        #here you can give a name to which each member of the iterable will be referd to
    print(fruit.upper())
```
